```python
# -*- coding: utf-8 -*-
"""1122_wordcloud_01_PSY.ipynb

Automatically generated by Colaboratory.

Original file is located at
    https://colab.research.google.com/drive/10N-2vmeQJMD47VJr8rjgEY47gRwc0E2R

# 1. AI 워드 클라우드
"""

# 워드 클라우드 위키피디아 사용을 위한 패키지 설치하기
!pip install wordcloud wikipedia
# 데이터 시각화를 위한 멧플롭립 패키지 설치하기
!pip install matplotlib

# 라이브러리 가져오기

# 위키피디아 사용을 위한 라이브러리
import wikipedia
# 그래프 및 시각화를 위한 라이브러리
import matplotlib.pyplot as plt          # as 별칭(내가 정하는 이름)
# 워드클라우드 생성을 위한 라이브러리
from wordcloud import WordCloud, STOPWORDS    # 중지어

# 위키피디아에서 컨텐츠 제목을 가져온다
wiki = wikipedia.page('Artificial intelligence')    # 제목을 가지고 와서 변수에 담는다

# 이 페이지의 텍스트 컨텐츠를 추출한다
text = wiki.content                                 # text 변수에 다 담는다

# 텍스트를 가지고 워드클라우드를 생성한다
wc = WordCloud(width=400, height=400).generate(text)      # 가로 400, 세로 400 텍스트를 가지고 워드클라우드를 만들어서 변수에 저장한다

# 새로운 figure를 생성한다 (이미지 생성)
plt.figure(figsize=(10, 10))    # 괄호 안에는 사이즈를 지정할 수 있음

# wc를 이미지로 표시해보자
plt.imshow(wc)  # wc의 이미지를 보여준다(이미지쇼), .imshow

# 그래프를 화면에 출력하자
plt.show()      # 화면에 이미지를 출력한다, .show

# 워드클라우드 저장하기
wc.to_file('AI_클라우드01.png')     # 이미지가 파일형태로 저장되어 있다, 괄호 안에는 파일명 적기, 다운로드 받을 수 있음

# 중지어가 제외된 워드클라우드 만들기 (필요없는 단어는 제외하자)
# 집합의 특징 : 중복을 허용하지 않는다 ==> union(교집합)
# 중복을 허용하지 않는 교집합의 특징을 활용하여 필요없는 단어들을 삭제해보자
sWords = STOPWORDS.union({'used', 'will', 'exanple', 'people', 'make', 'state', 'many', 'AI'})   # 교집합을 활용하여 필요없는 단어들을 삭제한다

# 텍스트를 가지고 워드클라우드를 생성한다
wc = WordCloud(width=400, height=400, stopwords = sWords).generate(text)      # sWord에 들어간 단어들은 제외하고 이미지를 만든다

# 새로운 figure를 생성한다 (이미지 생성)
plt.figure(figsize=(10, 10))    # 괄호 안에는 사이즈를 지정할 수 있음

# wc를 이미지로 표시해보자
plt.imshow(wc)  # wc의 이미지를 보여준다(이미지쇼), .imshow

# 그래프를 화면에 출력하자
plt.show()      # 화면에 이미지를 출력한다, .show

# 워드클라우드 저장하기
wc.to_file('AI_클라우드02.png')     # 이미지가 파일형태로 저장되어 있다, 괄호 안에는 파일명 적기, 다운로드 받을 수 있음
```
