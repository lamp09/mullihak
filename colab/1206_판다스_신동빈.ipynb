## 판다스의 DataFrame() 함수로 데이터프레임 만들기.

파이썬의 리스트ㅡ,파이썬의 딕셔너리, 넘파이의 배열을 활용.

리스트를 만들어 데이터 프레임의 데이터로 한 행씩 지정할 수 있음.

*pd.DataFrame(값, columns = 열 이름 리스트)*

```python
# 리스트를 만들어 데이터로 한 행씩 지정.
list1 = list([['홍길동','남자','20','180'],['김길동','남자','21','177'],['박길동','여자','20','160']])

# 열 이름 리스트 만들기
col_names = ['이름','성별','나이','키']
pd.DataFrame(list1, columns = col_names)
```

```python
# 딕셔너리로 데이터 프레임 만들기
# 딕셔너리 키는 열 이름, 값은 열에 대한 행의 데이터(리스트 형식 사용)

dict1 = {
    '이름' : {0 : '홍길동', 1 : '김길동', 2 : '박길동'},
    '성별' : {0 : '남자', 1 : '남자', 2 : '여자'},
    '나이' : {0 : '20', 1 : '21', 2 : '20'},
    '키'  : {0 : '180', 1 : '177', 2 : '160'}
}

pd.DataFrame(dict1)
```

```python
import pandas as pd
# 데이터 생성.
list1 = list([['홍길동','남자',30,183],
              ['지은탁','여자',24,162],
              ['한서준','남자',23,172],
              ['이수호','남자',21,183],
              ['임주경','여자',17,155],
              ['이두나','여자',26,157],
              ['강남순','여자',24,157],
              ['도봉순','여자',24,172]])

col_names = ['이름','성별','나이','키']

# 데이터 프레임으로 생성
df = pd.DataFrame(list1, columns = col_names)

# 출력
df
```

```python
# 나이가 21세 또는 23세인 사람을 조회.
df[df['나이'].isin([21,23])]
# 성별이 '여자'이면서 키가 160보다 큰 사람을 조회.
df[(df['성별'] == '여자') & (df['키'] > 160)]
```

```python
# 나이가 28세 이상이거나 성별이 '남자'인 사람을 조회.
df[(df['나이'] >= 28)|(df['성별'] == '남자')]
```

```python
# 이름에 '순'이 들어간 사람
df[df['이름'].str.contains('순')]
```

```python
# 성별의 평균 키
# 성별을 기준으로 그룹화하여 그룹별 키의 평균 계산
# as_index = False -> 데이터프레임으로 출력됨.
mean_by_gender = df.groupby(by = ['성별'], as_index= False)['키'].mean()

# 열 이름 바꾸기
mean_by_gender.rename(columns={'키' : '평균 키'}, inplace =True)

mean_by_gender
```

```python
# 성별의 키의 표준 편차
std_by_gender = df.groupby(by = ['성별'], as_index= False)['키'].std()

# 열 이름 바꾸기
std_by_gender.rename(columns={'키' : '키의 표준 편차'}, inplace =True)

std_by_gender
```

```python
# 두 데이터 프레임 합치기
new_df = pd.merge(mean_by_gender,std_by_gender)

new_df
```

```python
# 열 이름과 시간표 데이터를 리스트로 저장.
col_names = ['과목번호','과목명','강의실','시간수']
list1 = list((['C1','파이썬','303호',3],
              ['C2','국어','201호',3],
              ['C3','영어','203호',3],
              ['C4','한국사','201호',2],
              ['C5','과학','401호',3],
              ['C6','사회','201호',2]))
```

```python
# 시간표 데이터를 데이터프레임 객체 df로 변환하여 CSV파일로 저장
import pandas as pd

timetable_df = pd.DataFrame(list1,columns = col_names)

# CSV로 저장
timetable_df.to_csv('./timetable.csv', header = True, index = False, encoding ='utf-8')
```

```python
# CSV 파일 읽어와서 출력하기
timetable_df = pd.read_csv('./timetable.csv',sep =',')
timetable_df
```

```python
# 열 이름 '담당교사' 추가
# 값 추가 => '이순신', '김유신', '강감찬', '신사임당', '정약용', '허준'

timetable_df['담당교사'] = ['이순신', '김유신', '강감찬', '신사임당', '정약용', '허준']
timetable_df
```
